```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Diagrama 

x = np.arange(2, 20)
y = x**2
y2=x**3 
plt.scatter(x, y, color='red', label='x 2')
plt.scatter(x, y2, color='blue', label='x 3')

plt.title('Funcion de x  f(x) = x2')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()

plt.text(5,4000, 'TEXTO', color='black')
```
